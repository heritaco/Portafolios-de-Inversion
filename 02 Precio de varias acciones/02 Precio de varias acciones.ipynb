```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_multiple_stock_prices(tickers, start_date, end_date, colors, ticker_to_name):
    # Set Seaborn style
    sns.set(style="whitegrid", palette="pastel")

    # Create a figure and axis
    plt.figure(figsize=(14, 8))

    # Loop through each ticker and plot its stock price
    for ticker, color in zip(tickers, colors):
        # Download stock data from Yahoo Finance
        stock_data = yf.download(ticker, start=start_date, end=end_date)

        # Plot the closing prices
        sns.lineplot(data=stock_data, x=stock_data.index, y='Close', label=ticker_to_name[ticker], color=color, linewidth=2)

    # Customize the title and labels
    plt.title(f'Precios de acciones [{start_date} - {end_date}]', fontsize=18, fontweight='bold', color='black')
    plt.xlabel('Fecha', fontsize=14, fontweight='bold')
    plt.ylabel('Precio (USD)', fontsize=14, fontweight='bold')

    # Rotate the x-axis labels and set their font size
    plt.xticks(rotation=45, fontsize=11)

    # Customize y-axis ticks
    plt.yticks(fontsize=11)

    # Add a legend with a larger font size
    plt.legend(fontsize=13, title='Compañías', title_fontsize='13', loc='upper left', frameon=True)

    # Add vertical and horizontal grid lines
    plt.grid(True, linestyle='--', linewidth=0.7, axis='both')

    # Adjust the layout
    plt.tight_layout()

    # Display the plot
    plt.show()

# Example usage
tickers = ['NKE', 'ADDYY']
colors = ['blue', 'green']
ticker_to_name = {
    'NKE': 'Nike',
    'ADDYY': 'Adidas'
}
start_date = '2023-09-13'
end_date = '2024-09-21'

plot_multiple_stock_prices(tickers, start_date, end_date, colors, ticker_to_name)
```
