```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_daily_returns_histogram(ticker, start_date, end_date, color='blue'):
    # Set Seaborn style
    sns.set(style="whitegrid", palette="pastel")

    # Download stock data from Yahoo Finance
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    stock_data.reset_index(inplace=True)  # Reset index for easy plotting

    # Calculate daily returns
    stock_data['Daily Return'] = stock_data['Adj Close'].pct_change()

    # Create a figure and axis
    plt.figure(figsize=(14, 8))

    # Plot the histogram of daily returns
    sns.histplot(stock_data['Daily Return'].dropna(), bins=50, kde=True, color=color)

    # Customize the title and labels
    plt.title(f'Histograma de Rendimientos Diarios de {ticker} [{start_date} - {end_date}]', fontsize=18, fontweight='bold', color='black')
    plt.xlabel('Rendimiento Diario', fontsize=14, fontweight='bold')
    plt.ylabel('Frecuencia', fontsize=14, fontweight='bold')

    # Customize x-axis and y-axis ticks
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)

    # Add vertical and horizontal grid lines
    plt.grid(True, linestyle='--', linewidth=0.7, axis='both')

    # Adjust the layout
    plt.tight_layout()

    # Show the plot
    plt.show()

# Example usage
plot_daily_returns_histogram('NKE', '2023-09-13', '2024-09-13', color='#aaaaaa')
```
