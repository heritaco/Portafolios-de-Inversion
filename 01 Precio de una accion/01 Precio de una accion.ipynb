```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_stock_price(ticker, start_date, end_date, color, ticker_to_name):
    # Set Seaborn style
    sns.set(style="whitegrid", palette="pastel")

    # Download stock data from Yahoo Finance
    stock_data = yf.download(ticker, start=start_date, end=end_date)

    # Create a figure and axis
    plt.figure(figsize=(14, 8))

    # Plot the closing prices
    sns.lineplot(data=stock_data, x=stock_data.index, y='Close', label=ticker_to_name[ticker], color=color, linewidth=2)

    # Customize the title and labels
    plt.title(f'Precio de acción de {ticker_to_name[ticker]} [{start_date} - {end_date}]', fontsize=18, fontweight='bold', color='black')
    plt.xlabel('Fecha', fontsize=14, fontweight='bold')
    plt.ylabel('Precio (USD)', fontsize=14, fontweight='bold')

    # Rotate the x-axis labels and set their font size
    plt.xticks(rotation=45, fontsize=11)

    # Customize y-axis ticks
    plt.yticks(fontsize=11)

    # Add a legend with a larger font size
    plt.legend(fontsize=13, title='Compañía', title_fontsize='13', loc='upper left', frameon=True)

    # Add vertical and horizontal grid lines
    plt.grid(True, linestyle='--', linewidth=0.7, axis='both')

    # Adjust the layout
    plt.tight_layout()

    # Display the plot
    plt.show()
```

```python
# Example usage
ticker = 'NKE'
start_date = '2023-09-13'
end_date = '2024-09-13'
color = 'blue'
ticker_to_name = {
    'NKE': 'Nike',
    'ADDYY': 'Adidas',
    'UAA': 'Under Armour'
}

plot_stock_price(ticker, start_date, end_date, color, ticker_to_name)
```
